```python
import pandas as pd
import numpy as np

import utils as ut
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn import svm

from tqdm import tqdm

import seaborn as sns
from sklearn.metrics import confusion_matrix
import sklearn.metrics

from sklearn.model_selection import KFold
from sklearn.impute import KNNImputer

import pickle

from prettytable import PrettyTable
from prettytable import MSWORD_FRIENDLY

import time
#import tensorflow_addons as tfa
inicio = time.time()
```

```python
split_size = 10
```

```python
PATH = '../data/water_potability.csv'
label_column='Potability'
have_classes=True
drop_columns_list=['Id']
drop_columns=True
dropna = False

gamma_search = ['scale','auto']
C_search = range(1, 100, 2)
degree_search = range(1, 15, 1)

df, df_one_hot, X, y, classes, columns = ut.clean_data(PATH=PATH, shuffle=True, label_column=label_column, drop_columns_list=drop_columns_list, drop_columns=drop_columns, have_classes=have_classes, dropna=dropna)
df
```

```python
X =  preprocessing.scale(X, with_mean='False')
y = np.array(y)
X = np.array(X)

imputer = KNNImputer()
imputer.fit(X)
X = imputer.transform(X)
```

```python
def sort_func(item):
    return item[0]
```

```python
results = []
results.append(['acc', 'C'])

for j in tqdm(C_search):

    clf = svm.SVC(C=j/100, kernel='linear')

    clf.fit(X, y)

    predicoes = clf.predict(X)

    acc = sklearn.metrics.accuracy_score(y_true = y , y_pred = predicoes)

    results.append([acc, j/100])
        
print(results[0])
results = results[1:-1]
results.sort(key=sort_func, reverse=True)
results
```

```python
results = []
results.append(['acc', 'gamma', 'C'])

for i in tqdm(gamma_search):
    for j in C_search:

        clf = svm.SVC(C=j/100, gamma=i, kernel='rbf')

        clf.fit(X, y)
        
        predicoes = clf.predict(X)
        
        acc = sklearn.metrics.accuracy_score(y_true = y , y_pred = predicoes)
        
        results.append([acc, i, j/100])

print(results[0])
results = results[1:-1]
results.sort(key=sort_func, reverse=True)
results
```

```python
results = []
results.append(['acc', 'gamma', 'C', 'degree'])

for u in tqdm(degree_search):
    for i in gamma_search:
        for j in C_search:

            clf = svm.SVC(C=j/100, gamma=i, degree=u, kernel='poly')

            clf.fit(X, y)

            predicoes = clf.predict(X)

            acc = sklearn.metrics.accuracy_score(y_true = y , y_pred = predicoes)

            results.append([acc, i, j/100, u])

print(results[0])
results = results[1:-1]
results.sort(key=sort_func, reverse=True)
results
```
