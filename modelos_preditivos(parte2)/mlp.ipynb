```python
import tensorflow as tf
from tensorflow.keras import layers
import pandas as pd
import numpy as np
from tensorflow.keras import layers
from keras.callbacks import EarlyStopping
import utils as ut
import time
import tensorflow_addons as tfa
inicio = time.time()
```

```python
patience = 100
EPOCHS = 1500
batch_size = 1
save_model = True

t_aprendizado = 0.01
momentum = 0.01

PATH = '../data/water_potability.csv'
label_column='Potability'
have_classes=True
drop_columns_list=['Id']
drop_columns=True

df, df_one_hot, X, y, classes, columns = ut.clean_data(PATH=PATH, shuffle=True, label_column=label_column, drop_columns_list=drop_columns_list, drop_columns=drop_columns, have_classes=have_classes, dropna=True)
df
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10000, activation='relu'),
    tf.keras.layers.Dense(100000, activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='softmax')
  ])

#precision = tf.keras.metrics.Precision()
#recall = tf.keras.metrics.Recall()
#f1_score = tfa.metrics.F1Score(num_classes=1, threshold=0.5)

#opt=tf.keras.optimizers.SGD(learning_rate=t_aprendizado, momentum=momentum)
opt = 'adam'

model.compile(optimizer=opt,
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])

earlystopper = EarlyStopping(patience=patience, verbose=1)

model.fit (X, y, 
           validation_split=0.1, 
           batch_size=batch_size, epochs=EPOCHS, 
           callbacks=[earlystopper])
```

```python
if save_model:
    nome_saida = f'model.h5'
    #nome_saida = caminho_modelo+nome+'.h5'
    model.save(nome_saida)
    print('Salvo como: ',nome_saida)
print('Tempo de execução: ', time.time() - inicio)
```
